### 문제 1. 안전한 정수 입력 함수

사용자에게 숫자를 입력받는 함수 `input_int(prompt)`를 작성하세요.

1. 정수가 아니면 `"정수를 입력하세요."`를 출력하세요.
2. 올바른 정수를 입력할 때까지 반복하세요.
3. 정상 입력 시 `int` 값을 반환하세요.

```python
def input_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("정수를 입력하세요.")

# 호출
num = input_int("숫자를 입력하세요: ")
print(num)
```

### 문제 2. 나눗셈 계산기

두 정수를 입력받아 `a / b` 결과를 출력하세요.

1. 0으로 나누면 `"0으로 나눌 수 없습니다."`를 출력하세요.
2. 정수가 아닌 입력이면 `"정수만 입력하세요."`를 출력하세요.
3. 어떤 경우든 프로그램이 종료되지 않고 다시 입력받도록 작성하세요.

```python
def divide_calculator():
    while True:
        try:
            a = int(input("피제수: "))
            b = int(input("제수: "))
            
            result = a / b
            print(f"결과: {result}")
        
        except ValueError:
            print("정수만 입력하세요.")
            
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")
            
# 프로그램 시작 지점
if __name__ == "__main__":
    divide_calculator()
```

### 문제 3. 파일 읽기 예외 처리

`students.txt` 파일을 읽어 학생 점수의 평균을 출력하세요.

조건

1. 파일이 없으면 `"학생 데이터 파일이 없습니다. 먼저 생성하세요."`를 출력하세요.
2. 파일이 있으면 평균을 출력하세요.
3. 예외 처리는 `try / except / else` 구조로 작성하세요.

```python
try:
    with open("students.txt", "r", encoding="utf-8") as f:
        scores = []

        for line in f:
            name, score = line.strip().split(",")
            scores.append(int(score))

except FileNotFoundError:
    print("학생 데이터 파일이 없습니다. 먼저 생성하세요.")

else:
    avg = sum(scores) / len(scores)
    print(f"평균 점수: {avg}")
```

### 문제 4. 사용자 정의 예외로 규칙 강제

점수 입력 규칙: **0~100 사이 정수만 허용**

1. 규칙을 위반하면 `InvalidScoreError` 예외를 발생시키세요.
2. main 로직에서 해당 예외를 잡아 `"점수는 0~100 사이여야 합니다."`를 출력하세요.
3. 올바른 점수를 입력할 때까지 반복하세요.

```python
class InvalidScoreError(Exception):
    pass

def input_score():
    score = int(input("점수 입력: "))
    
    if score < 0 or score > 100:
        raise InvalidScoreError
    
    return score

# main 로직 (예외처리 + 반복)
while True:
    try:
        score = input_score()
        print(f"입력된 점수: {score}")
        break
        
    except ValueError:
        print("정수를 입력하세요.")
        
    except InvalidScoreError:
        print("점수는 0~100 사이여야 합니다.")
```

### 문제 5. finally 사용하기

프로그램 실행 중 예외가 발생할 수 있는 코드를 작성하세요.

조건

1. 예외가 발생하면 `"오류가 발생했습니다."`를 출력하세요.
2. 어떤 경우든 `"프로그램을 종료합니다."`가 마지막에 출력되도록 `finally`를 사용하세요.

```python
try:
    a = int(input("숫자 입력: "))
    result = 10 / a
    print(f"결과: {result}")

except:
    print("오류가 발생했습니다.")

finally:
    print("프로그램을 종료합니다.")
```
